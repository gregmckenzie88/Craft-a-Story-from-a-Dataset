# Adult Income Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('adult.csv')
df.replace('?', np.nan, inplace=True)
```

## Setup

My younger brother, Jim, has recently graduated from university with. Bachelor's degree.  Congrats!

## Conflict
After working in sales for a few months, Jim is entertaining the possibility that, if he goes back to school to earn a Master's degree, he will make more than 50K per year.  But he is not sure if this is worth the time, and the money it takes to achieve a higher level of education.

### Data
I have found a data source, the [Adult income dataset](https://www.kaggle.com/datasets/wenruliu/adult-income-dataset/code?datasetId=222&sortBy=voteCount) from Kaggle containing heaps of information all about income around the world!

Let's look at the 'income' column

```python
df['income'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
income_counts = df.groupby(['education', 'income']).size().unstack()
income_counts['ratio'] = income_counts['>50K'] / income_counts['<=50K']
sorted_income_ratios = income_counts['ratio'].sort_values(ascending=False)
sorted_income_ratios.plot(kind='bar', figsize=(10, 6))
plt.ylabel('Ratio (>50K / <=50K)')
plt.title('Income Ratio by Education')
plt.show()
```

#### How much more likley are you to earn over 50K if you have a Master's degree, rather than. Bachelor's?

```python
# Calculate the percentage for each education level and income class
perc = pd.crosstab(df['education'], df['income'], normalize='index') * 100

# Calculate the sum of percentages for '>50K' for each education level
# Assuming the '>50K' is one of the classes in 'income'
perc['sum'] = perc['>50K']

# Sort the DataFrame by the sum of percentages in descending order
perc = perc.sort_values('sum', ascending=False)

# Drop the 'sum' column as it is no longer needed
perc = perc.drop(columns=['sum'])

# Now, plot the sorted DataFrame
perc.plot(kind='bar', stacked=True)
plt.ylabel('Percentage')
plt.show()
```

```python
edu_filter = df[df['education'].isin(['Masters', 'Bachelors'])]
ct = pd.crosstab(edu_filter['education'], edu_filter['income'], normalize='index') * 100
ax = ct.plot(kind='barh', stacked=True, figsize=(10, 6))
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{width:.1f}%', (x + width/2, y + height/2), ha='center', va='center')

ax.set_xlabel('Percentage')
ax.set_title('Income Distribution for Masters and Bachelors')

plt.show()
```

### Education works! 

#### There is a total difference of 13.6%

But are there other ways to increase your income, aside form education?

```python
income_counts = df.groupby(['occupation', 'income']).size().unstack()
income_counts['ratio'] = income_counts['>50K'] / income_counts['<=50K']
sorted_income_ratios = income_counts['ratio'].sort_values(ascending=False)
sorted_income_ratios.plot(kind='bar', figsize=(10, 6))
plt.ylabel('Ratio (>50K / <=50K)')
plt.title('Income Ratio by Occupation')
plt.show()
```

It looks like Tech Support offers a higher ratio of over/under 50K than sales does.  Let's look further!

```python
edu_filter = df[df['occupation'].isin(['Sales', 'Tech-support'])]
ct = pd.crosstab(edu_filter['occupation'], edu_filter['income'], normalize='index') * 100
ax = ct.plot(kind='barh', stacked=True, figsize=(10, 6))
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{width:.1f}%', (x + width/2, y + height/2), ha='center', va='center')

ax.set_xlabel('Percentage')
ax.set_title('Income Distribution for Sales and Tech-support')

plt.show()
```

#### With a total difference of 2.2%, this does not beat getting a masters degree!
Let's try armed forces

```python
edu_filter = df[df['occupation'].isin(['Sales', 'Armed-Forces'])]
ct = pd.crosstab(edu_filter['occupation'], edu_filter['income'], normalize='index') * 100
ax = ct.plot(kind='barh', stacked=True, figsize=(10, 6))
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{width:.1f}%', (x + width/2, y + height/2), ha='center', va='center')

ax.set_xlabel('Percentage')
ax.set_title('Income Distribution for Sales and Armed-Forces')

plt.show()
```

#### With a diffference of only 6.5%, this is not a large jump!
Let's take a look at prof-specialty -- this looks like a large jump!

```python
edu_filter = df[df['occupation'].isin(['Sales', 'Prof-specialty'])]
ct = pd.crosstab(edu_filter['occupation'], edu_filter['income'], normalize='index') * 100
ax = ct.plot(kind='barh', stacked=True, figsize=(10, 6))
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{width:.1f}%', (x + width/2, y + height/2), ha='center', va='center')

ax.set_xlabel('Percentage')
ax.set_title('Income Distribution for Sales and Armed-Forces')

plt.show()
```

#### Although this is a larger jump than Armed Forces, it still does not break 50%!

### Instead of look at occupations, let's look at different sectors fo the economy

```python
perc = pd.crosstab(df['workclass'], df['income'], normalize='index') * 100
perc['sum'] = perc['>50K']
perc = perc.sort_values('sum', ascending=False)
perc = perc.drop(columns=['sum'])
perc.plot(kind='bar', stacked=True)
plt.ylabel('Percentage')
plt.show()
```

Let's take a closer look at `Self-emp-inc`

```python
edu_filter = df[df['workclass'].isin(['Self-emp-inc'])]
ct = pd.crosstab(edu_filter['workclass'], edu_filter['income'], normalize='index') * 100
ax = ct.plot(kind='barh', stacked=True, figsize=(10, 6))
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{width:.1f}%', (x + width/2, y + height/2), ha='center', va='center')

ax.set_xlabel('Percentage')
ax.set_title('Income Distribution for Self Employed -- Incorporated')
plt.show()
```

#### This is even higher than what we had for Master's degree

```python
self_emp_inc = df[df['workclass'] == 'Self-emp-inc']
education_counts = self_emp_inc['education'].value_counts(normalize=True) * 100
education_counts = education_counts.sort_values(ascending=True)
education_counts.plot(kind='barh')
plt.title('Education Distribution for Self-Employed Incorporated Individuals (%)')
plt.xlabel('Percentage (%)')
plt.ylabel('Education')
plt.show()
```

### Resolution

Will jim probably make more than 50K if he gets a Master's Degree.  Yes!

But that is not the only path forward.  Although, it seems switching occupations will likley not get him over the egde of 50K, becoming Self-Employed (Incorporated) will yield a slightly higher chance of making > 50K.  And the good news is this path does not appear to be restricted by education, as most people with this lifestyle are only high school graduates.
